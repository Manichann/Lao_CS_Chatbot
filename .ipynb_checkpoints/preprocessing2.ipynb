# Remove special characters

```python
filename = 'Data/bachelor_cs.txt'
cleaned_file = 'Cleaned_Data/cleaned_bachelor_cs.txt'
```

```python
def remove_special_characters(filename, cleaned_file):
    with open(filename, 'r', encoding='utf-8') as file:
        text = file.read()

    pattern = r'[^a-zA-Zກຂຄງຈສຊຍດຕຖທນບປຜຝພຟມຢລຫຼຣວຫອຮໜໝໆຽະາ ິ ີ ຶ ື ໂ ໍເແ ຸ ູຳໄໃ ັ ົ ່ ້ ໌]'  

    clean_text = re.sub(pattern, '', text)

    with open(cleaned_file, 'w', encoding='utf-8') as file:
        file.write(clean_text)

remove_special_characters(filename, cleaned_file)


def delete_space(cleaned_file):
    with open(cleaned_file, 'r',encoding='utf-8') as file:
        content = file.read()

    modifile_content = content.replace(" ", "")        

    with open(cleaned_file, 'w', encoding='utf-8') as file:
        file.write(modifile_content)

delete_space(cleaned_file)
```

```python
file_name = "corpus.txt"

existing_items = set()

# Read existing items from the file
try:
    with open(file_name, 'r', encoding='utf-8') as file:
        for line in file:
            # Remove newline characters and add to the set
            existing_items.add(line.strip())
except FileNotFoundError:
    # If the file does not exist, continue as there are no existing items
    pass

with open(file_name, 'a', encoding='utf-8') as file:
    for word in c:
        if word not in existing_items:
            file.write(word + '\n')
            existing_items.add(word)      
```
