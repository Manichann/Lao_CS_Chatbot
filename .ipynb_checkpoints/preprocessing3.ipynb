```python
from laonlp import word_tokenize
import re
```

```python
filename = '../Data/bachelor_web.txt'
```

```python
with open(filename, 'r', encoding='utf-8') as file:
    text = file.read()
```

# Remove special Characters

```python
def remove_special_character(text):
    return re.sub(r'[^ກຂຄງຈສຊຍດຕຖທນບປຜຝພຟມຢລຫຼຣວຫອຮໜໝໆຽະາ ິ ີ ຶ ື ໂ ໍເແ ຸ ູຳໄໃ ັ ົ ່ ້ ໌]', '', text)

def remove_spaces(text):
    
    return text.replace(" ", "") 
```

```python
clean_text = remove_special_character(text)
clean_text = remove_spaces(clean_text)
```

```python
print(clean_text)
```

# Word Tokenize

```python
tokens = word_tokenize(clean_text)
print(tokens)
```

```python
from collections import Counter

c = Counter(tokens)
len(c)
```

# Save to Lao Corpus

```python
from collections import Counter
c = Counter(tokens)
print(len(c),c)
```

```python
c = sorted(c)
print(c)
```

```python
file_name = "../lao_corpus.txt"

existing_items = set()

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        for line in file:
            existing_items.add(line.strip())
except FileNotFoundError:
    pass

with open(file_name, 'a', encoding='utf-8') as file:
    for word in c:
        if word not in existing_items:
            file.write(word + '\n')
            existing_items.add(word)  

print("save successfully")    
```
