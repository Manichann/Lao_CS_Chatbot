```python
from laonlp import word_tokenize
import re
```

```python
filename = 'Data/fns.txt'
cleaned_file = 'Cleaned_Data/cleaned_fns.txt'
```

# Clean Data

```python
def remove_special_characters(filename, cleanded_file):
    with open(filename, 'r', encoding='utf-8') as file:
        text = file.read()

    pattern = r'[^a-zA-Zກຂຄງຈສຊຍດຕຖທນບປຜຝພຟມຢລຫຼຣວຫອຮໜໝໆຽະາ ິ ີ ຶ ື ໂ ໍເແ ຸ ູຳໄໃ ັ ົ ່ ້ ໌]'
    clean_text = re.sub(pattern, '', text)

    with open(cleanded_file, 'w', encoding='utf-8') as file:
        file.write(clean_text)

remove_special_characters(filename, cleaned_file)


def delete_space(cleaned_file):
    with open(cleaned_file, 'r',encoding='utf-8') as file:
        content = file.read()

    modifile_content = content.replace(" ", "")        

    with open(cleaned_file, 'w', encoding='utf-8') as file:
        file.write(modifile_content)
        
delete_space(cleaned_file)
```

# Word Tokenize

```python
with open(cleaned_file, encoding='utf-8') as file:
    content = file.read()
    words = word_tokenize(content)
print(len(words), words)
```

# Re-Cleand Data

```python
split_values = ['ວິທະຍາສາດທຳມະຊາດ','FacultyOfScienceFOS','ຢູ່ໃນ','ສ້າງຕັ້ງຂຶ້ນ','ເລກວັດຖຸ','ກໍ່ໄດ້','ນັກສຶກສາວິຊາການຄົ້ນຄວ້າ',
                'ວິຊາວິທະຍາສາດ','LaosingaporeGeneticcomputercenter','FacultyofNaturalSciencesFNS','ຊີວະວິທະຍາວິທະຍາສາດ',
                'ວິຊາການເປັນ','ທີ່ມີຄຸນນະພາບ','ຄວາມຮູ້ຄວາມສາມາດ','ມີຄຸນສົມບັດ','ໃຫ້ມີມາດຕະຖານ','ບໍລິການທາງ','ວິຊາການທີ່ມີ',
                'ປົກປັກຮັກສາແລະ','ພະນັກງານໃຫ້','ໃນຫນ້າທີ່','ນັກສຶກສາກາຍເປັນ','ນັກວິຊາການທີ່ມີຄຸນສົມບັດ','ມີຄຸນນະພາບ','ພັດທະນາຄະນະ',
                'ໃຫ້ມີ','ໄດ້ທີ່']

items_to_insert = [['ວິທະຍາສາດ','ທຳມະຊາດ'],['Faculty','Of','Science','FOS'],['ຢູ່','ໃນ'],['ສ້າງຕັ້ງ','ຂຶ້ນ'],['ເລກ','ວັດຖຸ'],
                   ['ກໍ່','ໄດ້'],['ນັກສຶກສາ','ວິຊາການ','ຄົ້ນຄວ້າ'],['ວິຊາ','ວິທະຍາສາດ'],['Lao','singapore','Genetic',
                    'computer','center'],['Faculty','of','Natural','Sciences','FNS'],['ຊີວະວິທະຍາ','ວິທະຍາສາດ'],
                   ['ວິຊາການ','ເປັນ'],['ທີ່','ມີ','ຄຸນນະພາບ'],['ຄວາມຮູ້','ຄວາມສາມາດ'],['ມີ','ຄຸນສົມບັດ'],['ໃຫ້','ມີ','ມາດຕະຖານ'],
                   ['ບໍລິການ','ທາງ'],['ວິຊາການ','ທີ່','ມີ'],['ປົກປັກຮັກສາ','ແລະ'],['ພະນັກງານ','ໃຫ້'],['ໃນ','ຫນ້າທີ່'],
                   ['ນັກສຶກສາ','ກາຍເປັນ'],['ນັກວິຊາການ','ທີ່','ມີ','ຄຸນສົມບັດ'],['ມີ','ຄຸນນະພາບ'],['ພັດທະນາ','ຄະນະ'],
                   ['ໃຫ້','ມີ'],['ໄດ້','ທີ່']]

for split_value, items in zip(split_values, items_to_insert):
    index = 0
    while index < len(words):
        if words[index] == split_value:
            # Insert the items after the split value
            for item in items:
                words.insert(index + 1, item)
            # Move index past the newly inserted items
            index += len(items)
            words.remove(split_value)
        else:
            index += 1

print(len(words), words)
```

```python
items_to_remove = ['ປະລິນຍາ','ຕີ','ສົກ','ເຈ','ເນຕິກ','ເລີດ','ເຟັ້ນ','ຍອດ','ຄວາມ','httpfnsnuoledula']
items_to_add = ['ພາກວິຊາ','ປະລິນຍາຕີ','ສົກປີ','ເຈເນຕິກ','ເປັນເລີດ','ເລືອກເຟັ້ນ','ຍອດຍິ່ງ','ຄວາມທັນສະໄຫມ']

words = [item for item in words if item not in items_to_remove]
words.extend(items_to_add)
print(len(words), words)
```

# Save to Lao Corpus

```python
from collections import Counter
c = Counter(words)
print(len(c),c)
```

```python
c = sorted(c)
print(c)
```

```python
file_name = "corpus.txt"

existing_items = set()

# Read existing items from the file
try:
    with open(file_name, 'r', encoding='utf-8') as file:
        for line in file:
            # Remove newline characters and add to the set
            existing_items.add(line.strip())
except FileNotFoundError:
    # If the file does not exist, continue as there are no existing items
    pass

with open(file_name, 'a', encoding='utf-8') as file:
    for word in c:
        if word not in existing_items:
            file.write(word + '\n')
            existing_items.add(word)  

print("save successfully")   
```
