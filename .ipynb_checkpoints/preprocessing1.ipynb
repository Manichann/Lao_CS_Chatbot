# Clean Data

```python
def remove_special_characters(filename, cleanded_file):
    with open(filename, 'r', encoding='utf-8') as file:
        text = file.read()

    pattern = r'[^a-zA-Zກຂຄງຈສຊຍດຕຖທນບປຜຝພຟມຢລຫຼຣວຫອຮໜໝໆຽະາ ິ ີ ຶ ື ໂ ໍເແ ຸ ູຳໄໃ ັ ົ ່ ້ ໌]'
    clean_text = re.sub(pattern, '', text)

    with open(cleanded_file, 'w', encoding='utf-8') as file:
        file.write(clean_text)

remove_special_characters(filename, cleaned_file)


def delete_space(cleaned_file):
    with open(cleaned_file, 'r',encoding='utf-8') as file:
        content = file.read()

    modifile_content = content.replace(" ", "")        

    with open(cleaned_file, 'w', encoding='utf-8') as file:
        file.write(modifile_content)
        
delete_space(cleaned_file)
```

```python
def delete_text_from_file(cleaned_file, text_to_delete):
    with open(cleaned_file, 'r',encoding='utf-8') as file:
        content = file.read()

    modified_content = content.replace(text_to_delete, '')

    with open(cleaned_file, 'w', encoding='utf-8') as file:
        file.write(modified_content)

text_to_delete = "REສະຖິຕິວິໄຈພື້ນຖານທົ່ວໄປDSDataScienceວິຊາສະເພາະMBການຂຽນໂປຣແກຣມສຳລັບມືຖືພື້ນຖານວິຊາສະເພາະOSລະບົບປະຕິບັດການພື້ນຖານວິຊາສະເພາະOOການພັດທະນາລະບົບດ້ວຍຮູບແບບວັດຖຸພື້ນຖານວິຊາສະເພາະADໂຄງສ້າງຂໍ້ມູນແລະຂັ້ນຕອນວິທີວິຊາສະເພາະMIລະບົບຂ່າວສານເພື່ອການບໍລິຫານພື້ນຖານທົ່ວໄປລວມປີທີພາກຮຽນທີລຳດັບລະຫັດວິຊາຊື່ວິຊາໜ່ວຍກິດໝວດວິຊາDMການຂຸດຄົ້ນຂໍ້ມູນDataMiningວິຊາສະເພາະMCMathematicsforComputingພື້ນຖານທົ່ວໄປCNການສື່ສານຂໍ້ມູນແລະເຄືອຂ່າຍພື້ນຖານວິຊາສະເພາະDBລະບົບຖານຂໍ້ມູນພື້ນຖານວິຊາສະເພາະDDDesktopApplicationDevelopmentວິຊາສະເພາະSAການວິເຄາະແລະອອກແບບລະບົບວິຊາສະເພາະລວມປີທີພາກຮຽນທີລຳດັບລະຫັດວິຊາຊື່ວິຊາໜ່ວຍກິດໝວດວິຊາSMສຳມະນາພື້ນຖານວິຊາສະເພາະRMວິທີການຄົ້ນຄວ້າວິຊາສະເພາະSEວິສະວະກຳຊອບແວວິຊາສະເພາະDAການຄຸ້ມຄອງຖານຂໍ້ມູນວິຊາສະເພາະWSWebServerandTechnologyວິຊາສະເພາະCRການສ້າງລາຍງານດ້ວຍCrystalReportວິຊາສະເພາະWPWebProgrammingວິຊາສະເພາະລວມປີທີພາກຮຽນທີລຳດັບລະຫັດວິຊາຊື່ວິຊາໜ່ວຍກິດໝວດວິຊາCSຄວາມປອດໄພແລະຈັນຍາບັນການນຳໃຊ້ຄອມພິວເຕີພື້ນຖານວິຊາສະເພາະAIປັນຍາປະດິດພື້ນຖານວິຊາສະເພາະXXXXXXXXXວິຊາເລືອກFPໂຄງການຈົບຊັ້ນວິຊາສະເພາະລວມ"
delete_text_from_file(cleaned_file, text_to_delete)
```

```python
split_values = ['ContinuingEducationProgramtoBachelorofScienceinComputerScience',
                 'ContinuingEducationProgramtoBachelorofScienceinComputerScienceContBScComputerScience',
                 'ວິທະຍາສາດທຳມະຊາດ','ສາຂາວິທະຍາສາດ','ຕນປຕວ','ສ້າງຂຶ້ນ','ກະຊວງສຶກສາທິການ','ທີ່ມີຄຸນນະພາບ','ໃຫ້ມີຄຸນນະທຳ',
                 'ນັກວິຊາການໃຫ້ມີ','ວິທີການຄົ້ນຄວ້າ','ສາຂາວິຊາວິທະຍາສາດ','ເວລາການເປີດ','ທີ່ຈະ','ການສຶກສາລະດັບ','ສາຂາວິຊາ',
                 'ປີການສຶກສາ','ນັກສຶກສາປີ','ວິທີການເລືອກ','ຄະນະກຳມະການສອບເສັງ','ແຕ່ລະພາກ','ກິດການກຳນົດການຈັດຕັ້ງ',
                 'ຫາວັນສຸກ','ນີ້ມີການຈັດຕັ້ງ','ທັດສະນະສຶກສາ','ເວລາການສຶກສານັກສຶກສາ','ນີ້ມີ','ໂຄງການຈົບຊັ້ນ','ແຜນການຮຽນ','ໃນດ້ານ']

items_to_insert = [['Continuing','Education','Program','to','Bachelor','of','Science','in','Computer','Science'],
                   ['Continuing','Education','Program','to','Bachelor','of','Science','in','Computer','Science',
                    'Cont.BSc','Computer','Science'],['ວິທະຍາສາດ','ທຳມະຊາດ'],['ສາຂາ','ວິທະຍາສາດ'],['ຕນ.ປຕ.ວ'],
                   ['ສ້າງ','ຂຶ້ນ'],['ກະຊວງ','ສຶກສາທິການ'],['ທີ່','ມີ','ຄຸນນະພາບ'],['ໃຫ້','ມີ','ຄຸນນະທຳ'],
                   ['ນັກວິຊາການ','ໃຫ້','ມີ'],['ວິທີ','ການຄົ້ນຄວ້າ'],['ສາຂາ','ວິຊາ','ວິທະຍາສາດ'],['ເວລາ','ການເປີດ'],['ທີ່','ຈະ'],
                   ['ການສຶກສາ','ລະດັບ'],['ສາຂາ','ວິຊາ'],['ປີ','ການສຶກສາ'],['ນັກສຶກສາ','ປີ'],['ວິທີ','ການເລືອກ'],
                   ['ຄະນະ','ກຳມະການ','ສອບເສັງ'],['ແຕ່ລະ','ພາກ'],['ກິດ','ການກຳນົດ','ການຈັດຕັ້ງ'],['ຫາ','ວັນສຸກ'],
                   ['ນີ້','ມີ','ການຈັດຕັ້ງ'],['ທັດສະນະ','ສຶກສາ'],['ເວລາ','ການສຶກສາ','ນັກສຶກສາ'],['ນີ້','ມີ'],['ໂຄງການ','ຈົບຊັ້ນ'],
                   ['ແຜນ','ການຮຽນ'],['ໃນ','ດ້ານ']]

for split_value, items in zip(split_values, items_to_insert):
    index = 0
    while index < len(words):
        if words[index] == split_value:
            # Insert the items after the split value
            for item in items:
                words.insert(index + 1, item)
            # Move index past the newly inserted items
            index += len(items)
            words.remove(split_value)
        else:
            index += 1
```

```python
items_to_remove = ['ປະລິນຍາ','ຕີ','ພາກ','ວິຊາວິທະຍາສາດ','ຫຼັກ','ມູນ','ຄວາມ','ນັກ','ໆ','ຈາ', 'ລິ','ຍະທຳ','ສົກ','ອະນຸ',
                   'ດັ່ງ','ໜ່ວຍ','ກິດ']
items_to_add = ['ປະລິນຍາຕີ','ພາກວິຊາ','ວິທະຍາສາດ','ຫຼັກມູນ','ຄວາມຮຽກຮ້ອງ','ນັກຄົ້ນຄວ້າ','ໃໝ່ໆ','ຈະລິຍະທຳ','ສົກຮຽນ',
                'ອະນຸປະລິນຍາ','ພາກຮຽນ','ດັ່ງນີ້','ໜ່ວຍກິດ']

words = [item for item in words if item not in items_to_remove]
words.extend(items_to_add)

print(len(words), words)
```
