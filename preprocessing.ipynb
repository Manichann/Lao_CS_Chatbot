# 2. Remove special characters

```python
def remove_special_characters(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        text = file.read()

    # Define the pattern for special characters using regular expressions
    # Lao script ranges from U+0E80 to U+0EFF in Unicode
    pattern = r'[^a-zA-Z0-9ກຂຄງຈສຊຍດຕຖທນບປຜຝພຟມຢລຫຼວຫອຮໜໝໆຽະາ ິ ີ ຶ ື ໂ ໍເແ ຸ ູຳໄໃ ັ ົ ່ ້ ]'  # Keep Lao characters and whitespaces

    # Remove special characters from the text using the pattern
    clean_text = re.sub(pattern, '', text)

    # Write the cleaned text back to the output file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(clean_text)

# Example usage:
input_file = 'Data/bachelor-cs.txt'
output_file = 'Data/cleaned-bachelor-cs.txt'
remove_special_characters(input_file, output_file)
```
