```python
from laonlp import word_tokenize, lao_stopwords
import pandas as pd
import re
```

```python
filename = pd.read_csv('QnA.csv')

q = filename.iloc[:, 0].values
a = filename.iloc[:, 1].values
```

# Create Vocabulary

```python
word2count = {}

question = [word_tokenize(i) for i in q]
for i, tokens in enumerate(question):
    for word in tokens:
        if word not in word2count:
            word2count[word] = 1
        else: 
            word2count[word] += 1    

answer = [word_tokenize(i) for i in a]
for i, tokens in enumerate(answer):
    for word in tokens:
        if word not in word2count:
            word2count[word] = 1
        else: 
            word2count[word] += 1 
```

```python
print(len(word2count), word2count)
```

```python
vocab = {}
word_num = 1
for word, count in word2count.items():
    vocab[word] = word_num
    word_num += 1

vocab
```

```python
tokens = ['<PAD>', '<EOS>', '<OUT>', '<SOS>']
x = len(vocab)
for token in tokens:
    x += 1
    vocab[token] = x
```

```python
vocab['<PAD>'] = 0
```

```python
inv_vocab = {w:v for v, w in vocab.items()}
inv_vocab
```

# Clean Question

```python
def remove_special_character(text):
    return re.sub(r'[^0-9a-zA-Zກຂຄງຈສຊຍດຕຖທນບປຜຝພຟມຢລຫຼຣວຫອຮໜໝໆຽະາ ິ ີ ຶ ື ໂ ໍເແ ຸ ູຳໄໃ ັ ົ ່ ້ ໌]', '', text)
    
clean_q = [remove_special_character(text) for text in q]

def remove_spaces(text):
    
    return text.replace(" ", "")

clean_q = [remove_spaces(text) for text in q]

clean_q
```

```python
stop_words = ['ແດ່', 'ໃຜ', 'ເປັນ','ແລ້ວ', 'ແນ່','ມາ','ຖ້າ','ແມ່ນ','ຂໍ','ເບິ່ງ','ຂອງ','ຈະ','ໃນ','ເທົ່າໃດ']
```

```python
question = [word_tokenize(i)for i in clean_q]
for i, tokens in enumerate(question):
    stop_word = lao_stopwords()
    words = [w for w in tokens if w not in stop_word]

    dataX.append(words)
```

```python
from keras.models import Model
from keras.layers import Dense, Embedding, LSTM, Input

enc_inp = Input(shape=(15, ))
dec_inp = Input(shape=(100, ))
```

```python
VOCAB_SIZE = len(vocab) + 1
embedding_dim = 256

enc_embed = Embedding(VOCAB_SIZE, embedding_dim)(enc_inp)
enc_lstm = LSTM(512, return_state=True)
enc_op, h, c = enc_lstm(enc_embed)
enc_states = [h, c]
```

```python
dec_embed = Embedding(VOCAB_SIZE, embedding_dim)(dec_inp)
dec_lstm = LSTM(512, return_sequences=True, return_state=True)
dec_op, _, _ = dec_lstm(dec_embed, initial_state=enc_states)
```

```python
dense = Dense(VOCAB_SIZE, activation='softmax')
dec_op = dense(dec_op)
```

```python
enc_model = Model(enc_inp, enc_states)
```

```python
decoder_state_input_h = Input(shape=(512,))
decoder_state_input_c = Input(shape=(512,))

decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_lstm_outputs, state_h, state_c = dec_lstm(dec_embed, initial_state=decoder_states_inputs)

decoder_states = [state_h, state_c]
decoder_outputs = dense(decoder_lstm_outputs)
dec_model = Model([dec_inp] + decoder_states_inputs, [decoder_outputs] + decoder_states)
```

```python
def extract_keywords(input_text, dataX):
    keywords = []
    for keyword in dataX:
        if keyword in input_text:
            keywords.append(keyword)
    return keywords
```

```python
predefined_keywords = []
for i in dataX:
    for word in i:
        if word not in predefined_keywords:
            predefined_keywords.append(word)

predefined_keywords
```

```python
def decode_sequence(input_seq):
    # Encode the input as state vectors.

    states_value = enc_model.predict(input_seq)

    # Generate empty target sequence of length 1.
    target_seq = np.zeros((1,1))
    target_seq[0,0] = vocab['<SOS>']

    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        output_tokens, h, c = dec_model.predict(
            [target_seq] + states_value)

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = inv_vocab[sampled_token_index]
        decoded_sentence += '' + sampled_word

        # Exit condition: either hit max length or find stop token.
        if (sampled_word == '<EOS>' or len(word_tokenize(decoded_sentence)) > 50):
            stop_condition = True

        # Update the target sequence (length 1).
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        # Update states
        states_value = [h, c]

    return decoded_sentence
```

```python
def decode_sequence(input_seq, predefined_keywords):
    # Encode the input as state vectors.

    input_keywords = extract_keywords(input_seq, predefined_keywords)
    if not input_keywords:
        return "Sorry, I don't understand."

    states_value = enc_model.predict(input_seq)

    # Generate empty target sequence of length 1.
    target_seq = np.zeros((1,1))
    target_seq[0,0] = vocab['<SOS>']

    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        output_tokens, h, c = dec_model.predict(
            [target_seq] + states_value)

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = inv_vocab[sampled_token_index]
        decoded_sentence += '' + sampled_word

        # Exit condition: either hit max length or find stop token.
        if (sampled_word == '<EOS>' or len(word_tokenize(decoded_sentence)) > 50):
            stop_condition = True

        # Update the target sequence (length 1).
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        # Update states
        states_value = [h, c]

    return decoded_sentence if decoded_sentence else "Sorry, I don't understand that."
```

```python
while True:
    input_text = input("You: ")
    if input_text == 'exit':
        break
    input_seq = remove_special_character(input_text)
    input_seq = remove_spaces(input_seq)
    txt = []
    lst = []
    input_seq = word_tokenize(input_seq)
    stop_word = lao_stopwords()
    words = [w for w in input_seq if w not in stop_word]
    for x in words:
        try:
            lst.append(vocab[x])
        except:
            lst.append(vocab['<OUT>'])
    txt.append(lst)
    
    input_padded_seq = pad_sequences(txt, 15, padding='post', truncating='post')
    response = decode_sequence(input_padded_seq)
    print(f"You: {input_text} / {input_padded_seq}")
    print(f"Bot: {response}")
```
